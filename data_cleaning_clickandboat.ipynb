```python
# leer fichero Scrapping_final_2026-07-24_26.xlsx  
import pandas as pd
df = pd.read_excel("Scrapping_final_2026-07-24_26.xlsx")
```

```python
#Para aquellas celdas vacías en tipo_barco ver si se puede extraer palabras clave de la URL y rellenar esos valores en tipo_barco.
import numpy as np
def extract_boat_type(url):
    if pd.isnull(url):
        return np.nan
    url = url.lower()
    if 'sailboat' in url or 'velero' in url:
        return 'Velero'
    elif 'motorboat' in url or 'motor' in url:
        return 'Motorboat'
    elif 'catamaran' in url:
        return 'Catamarán'
    elif 'yacht' in url or 'yate' in url:
        return 'Yate'
    else:
        return np.nan
df['tipo_barco'] = df.apply(lambda row: extract_boat_type(row['url_barco']) if pd.isnull(row['tipo_barco']) else row['tipo_barco'], axis=1)

```

```python
#ver si tengo algun nulo en tipo_barco
df[df['tipo_barco'].isnull()]
```

```python
#Para aquellas celdas vacías en tipo_barco ver si se puede extraer palabras clave de nombre_barco
def extract_boat_type_name(name):
    if pd.isnull(name):
        return np.nan
    name = name.lower()
    if 'sailboat' in name or 'velero' in name:
        return 'Velero'
    elif 'motorboat' in name or 'motor' in name:
        return 'Motorboat'
    elif 'catamaran' in name:
        return 'Catamarán'
    elif 'yacht' in name or 'yate' in name:
        return 'Yate'
    else:
        return np.nan
df['tipo_barco'] = df.apply(lambda row: extract_boat_type_name(row['nombre_barco']) if pd.isnull(row['tipo_barco']) else row['tipo_barco'], axis=1)
```

```python
#ver si tengo algun nulo en tipo_barco
df[df['tipo_barco'].isnull()]
```

```python
#exportar en excel
df.to_excel("data_cleaning_clickandboat_2026-07-24_26.xlsx", index=False)
print("DataFrame limpio guardado como 'data_cleaning_clickandboat_2026-07-24_26.xlsx'")
```

```python
#Si la URL contiene "sin-licencia" rellenar "motorboat" tenga o no datos rellenos en tipo_barco
df.loc[df['url_barco'].str.contains('sin-licencia', na=False), 'tipo_barco'] = 'Motorboat'
#ver si tengo algun nulo en tipo_barco
df[df['tipo_barco'].isnull()]
```

```python
#muestrame los tipos_barco y su conteo
df['tipo_barco'].value_counts()
```

```python
#motorboat y lancha son lo mismo asi que quiero poner lancha
df['tipo_barco'] = df['tipo_barco'].replace('Motorboat', 'Lancha')
#muestrame los tipos_barco y su conteo
df['tipo_barco'].value_counts()
```

```python
#añadir una nueva columna al df que sea el nombre de la ciudad y que se saque de origen_fichero columna
#la ciudad aparece después de "clickandboat_"
df['ciudad'] = df['origen_fichero'].str.extract(r'clickandboat_(\w+)_\d{4}-\d{2}-\d{2}_\d{2}')
df.head()
```

```python
#precio borrar el icono del € y pasarlo a numérico
df['precio_dia'] = (
    df['precio_dia']
      .str.replace('€', '', regex=False)
      .str.replace('\xa0', '', regex=False)
      .str.replace(' ', '', regex=False)
      .str.replace('.', '', regex=False)
      .str.replace(',', '.', regex=False)
)
df['precio_dia'] = pd.to_numeric(df['precio_dia'], errors='coerce')  # los raros se vuelven NaN
df.head()
```

```python
#saber el tipo de dato para todas las columnas
df.dtypes
```

```python
#en el campo "capacidad_personas" borrar la palabra "personas" en un campo que es tipo object y pasarlo a numérico
df['capacidad_personas'] = df['capacidad_personas'].str.replace(' personas', '')
#pasar capacidad_personas a numérico
df['capacidad_personas'] = pd.to_numeric(df['capacidad_personas'], errors='coerce')
df.head()
```

```python
df.dtypes
```

```python
#mostrar el recuento de filas
len(df)
```

```python
#borrar las filas que tengan precio_dia vacío
df = df.dropna(subset=['precio_dia'])
```

```python
#mostrar el recuento de filas
len(df)
```

```python
#limpiar "eslora_m" porque tiene un caracter raro al principio y la m del final la quiero quitar también
df['eslora_m'] = df['eslora_m'].str.replace(r'[^0-9.]', '', regex=True)
#pasar eslora_m a numérico
df['eslora_m'] = pd.to_numeric(df['eslora_m'], errors='coerce') 
df.head()
```

```python
df.dtypes
```

```python
#Si en url contiene sin-licencia poner en columna patron: "No patrón. Sin licencia"
df.loc[df['url_barco'].str.contains('sin-licencia', na=False), 'patron'] = 'No patrón. Sin licencia'
df.head()
```

```python
#nombra los numero de registros que tenemos por zona
df['ciudad'].value_counts()
```

```python
#quiero que me lista la zona que hay en cada ciudad, zona y ciudad son dos columnas
df[['ciudad', 'zona']].drop_duplicates().sort_values(by=['ciudad', 'zona'])
```

```python
#borrar la linea que contenga en zona las siguientes zona: Aguadulce, La Línea de la Concepción, Algeciras, Puerto de Sotogrande, San Roque, Almuñécar, Gibraltar, La Alcaidesa, La Herradura, Sotogrande, Torreguadiario
zonas_a_borrar = ['Aguadulce', 'La Línea de la Concepción', 'Algeciras', 'Puerto de Sotogrande', 'San Roque', 'Almuñécar', 'Gibraltar', 'La Alcaidesa', 'La Herradura', 'Sotogrande', 'Torreguadiario']
df = df[~df['zona'].isin(zonas_a_borrar)] 
#mostrar recuento por filas
df.shape[0]
```

```python
#recuento de numero de barcos por ciudad
df['ciudad'].value_counts()
```

```python
#hay un nombre_barco que se llama Lancha 8 pax quiero cambiar el tipo_barco a Lancha y sobrescribir valor actual
df.loc[df['nombre_barco'] == 'Lancha 8 pax', 'tipo_barco'] = 'Lancha'
```
